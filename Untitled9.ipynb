```python
#when was the higest energy consumption and which year
df[df["AEP_MW"]==df["AEP_MW"].max()]
```

```python
#plot and data visualization
sns.distplot(df["AEP_MW"])
```

```python
#extract year from datetime
df["year"]=pd.DatetimeIndex(df['Datetime']).year
```

```python
df.head()
```

```python
#check how many year are unique
df["year"].unique()
```

```python
#Relationship of Energy vs year
sns.lineplot(x=df["year"],y=df["AEP_MW"],data=df)
```

```python
#Regression
sns.jointplot(x=df["year"],y=df["AEP_MW"],data=df,kind="reg")
```

```python
sns.jointplot(x=df["year"],y=df["AEP_MW"],data=df,kind="kde")
```
